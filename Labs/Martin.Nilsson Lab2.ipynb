## Samband mellan olika faktorer och examensgrad

Nedan tog jag fram fyra histogram som jÃ¤mfÃ¶r examensgrad "Grad.Rate" mellan olika grupper av skolor. 
Jag ville se om skoltyp, alumniengagemang, lÃ¤rartÃ¤thet och resurser per student pÃ¥verkar examensgraden. Logiskt redan nu bÃ¶r t.ex privata skolor ha hÃ¶gre examensgrad, har lÃ¤rarna fÃ¤rre studenter att ta hand om bÃ¶r vi ocksÃ¥ se hÃ¶gre examensgrad, bara fÃ¶r att nÃ¤mna tvÃ¥ tidiga tankar om det hela. 

SÃ¥ nedan ser vi jÃ¤mfÃ¶relser enligt detta. 
Varje subplot visar en jÃ¤mfÃ¶relse:
- **Ã–vre vÃ¤nster:** Private vs Public skolor  
- **Ã–vre hÃ¶ger:** HÃ¶g vs LÃ¥g andel aktiva alumner  
- **Nedre vÃ¤nster:** LÃ¥g vs HÃ¶g studentâ€“faculty ratio

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 5))
college[college["Private"] == "Yes"]["Grad.Rate"].plot.hist(bins = 20, ax=axes[0,0], color="skyblue", edgecolor="white",label="Private", alpha=0.5)
college[college["Private"] == "No"]["Grad.Rate"].plot.hist(bins = 20, ax=axes[0,0], color="salmon", edgecolor="white",label="Public", alpha=0.5)
axes[0,0].set_title("Private/Public - Graduation rate")
axes[0,0].legend()

college["Alumni_Group"] = pd.cut(college["perc.alumni"], bins=[0,21,100], labels=["Low alumni", "High alumni"])
college[college["Alumni_Group"]=="Low alumni"]["Grad.Rate"].plot.hist(bins = 20 ,ax=axes[0,1], color="skyblue", edgecolor="white",label="Low alumni", alpha=0.5)
college[college["Alumni_Group"]=="High alumni"]["Grad.Rate"].plot.hist(bins = 20 ,ax=axes[0,1], color="salmon", edgecolor="white",label="High alumni", alpha=0.5)
axes[0,1].set_title("High/low alumni - Graduation rate")
axes[0,1].legend()


college["SFR_Group"] = pd.cut(college["S.F.Ratio"], bins=[0,13, college["S.F.Ratio"].max()], labels=["Low ratio (â‰¤10)", "High ratio (>10)"])
college[college["SFR_Group"]=="Low ratio (â‰¤10)"]["Grad.Rate"].plot.hist(bins = 20 ,ax=axes[1,0], color="skyblue", edgecolor="white",label="Low Ratio", alpha=0.5)
college[college["SFR_Group"]=="High ratio (>10)"]["Grad.Rate"].plot.hist(bins = 20 ,ax=axes[1,0], color="salmon", edgecolor="white",label="High Ratio", alpha=0.5)
axes[1,0].set_title("Studentâ€“Faculty Ratio â€“ Graduation rate")
axes[1,0].legend()


college["Expend_Group"] = pd.cut(college["Expend"], bins=[0,10000, college["Expend"].max()], labels=["Low expend", "High expend"])
college[college["Expend_Group"]=="Low expend"]["Grad.Rate"].plot.hist(bins = 20 ,ax=axes[1,1], color="skyblue", edgecolor="white",label="Low expenditure", alpha=0.5)
college[college["Expend_Group"]=="High expend"]["Grad.Rate"].plot.hist(bins = 20 ,ax=axes[1,1], color="salmon", edgecolor="white",label="High expenditure",alpha=0.5)
axes[1,1].set_title("Expenditures per Student â€“ Graduation rate")
axes[1,1].legend()

for ax in axes.flatten():
    ax.set_xlabel("Graduation Rate (%)")
    ax.set_ylabel("Number of Colleges")

fig.suptitle("Graduation Rate Distributions by Different Factors", fontsize=14)

plt.tight_layout()
plt.show()
```
